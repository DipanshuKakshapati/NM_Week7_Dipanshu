### Analytical

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the analytical solution
def analytical_solution(t):
    return np.exp(0.5 * t**2 - 1.5 * t)

# Create a time vector from 0 to 2
t = np.linspace(0, 2, 100)

# Compute the analytical solution
y_analytical = analytical_solution(t)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(t, y_analytical, label='Analytical Solution', color='blue')
plt.title('Solution of the Differential Equation')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.grid(True)
plt.show()
```

### Eulers Method with h = 0.5 & h = 0.25

```python
# Define the function f(t, y) = yt - 1.5y
def f(t, y):
    return y * t - 1.5 * y

# Implement Euler's method
def euler_method(h, t_final):
    t_values = np.arange(0, t_final + h, h) # t values from 0 to t_final with step h
    y_values = np.zeros(len(t_values)) # Initialize y values array
    y_values[0] = 1 # Initial condition
    
    for i in range(1, len(t_values)):
        y_values[i] = y_values[i-1] + h * f(t_values[i-1], y_values[i-1])
    
    return t_values, y_values

# Euler's method with h = 0.5
t_05, y_05 = euler_method(0.5, 2)

# Euler's method with h = 0.25
t_025, y_025 = euler_method(0.25, 2)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(t, y_analytical, label='Analytical Solution', color='blue')
plt.plot(t_05, y_05, 'o-', label="Euler's Method, h=0.5", color='red')
plt.plot(t_025, y_025, 's-', label="Euler's Method, h=0.25", color='green')
plt.title('Solution of the Differential Equation using Euler’s Method')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the differential equation
def dy_dt(t, y):
    return y*t - 1.5*y

# Analytical solution function
def analytical_solution(t):
    return np.exp(0.5*t**2 - 1.5*t)

# Euler's method function
def euler_method(h, t_final):
    t_values = np.arange(0, t_final + h, h)
    y_values = np.zeros(len(t_values))
    y_values[0] = 1  # Initial condition
    for i in range(1, len(t_values)):
        y_values[i] = y_values[i-1] + h*dy_dt(t_values[i-1], y_values[i-1])
    return t_values, y_values

# Step sizes
h1 = 0.5
h2 = 0.25

# Calculate solutions
t_analytical = np.linspace(0, 2, 100)
y_analytical = analytical_solution(t_analytical)
t_euler1, y_euler1 = euler_method(h1, 2)
t_euler2, y_euler2 = euler_method(h2, 2)

# Plotting
plt.figure(figsize=(10, 6))
plt.plot(t_analytical, y_analytical, label='Analytical Solution', color='blue')
plt.plot(t_euler1, y_euler1, 'o-', label=f"Euler's Method, h={h1}", color='red')
plt.plot(t_euler2, y_euler2, 's-', label=f"Euler's Method, h={h2}", color='green')
plt.title('Solving Differential Equation with Euler’s Method')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.grid(True)
plt.show()
```

### Midpoint method with h = 0.5

```python
# Define the midpoint method function
def midpoint_method(h, t_final):
    t_values = np.arange(0, t_final + h, h) # t values from 0 to t_final with step h
    y_values = np.zeros(len(t_values)) # Initialize y values array
    y_values[0] = 1 # Initial condition
    
    for i in range(1, len(t_values)):
        k1 = f(t_values[i-1], y_values[i-1])
        y_mid = y_values[i-1] + (h/2) * k1
        t_mid = t_values[i-1] + h/2
        k2 = f(t_mid, y_mid)
        y_values[i] = y_values[i-1] + h * k2
    
    return t_values, y_values

# Midpoint method with h = 0.5
t_midpoint, y_midpoint = midpoint_method(0.5, 2)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(t, y_analytical, label='Analytical Solution', color='blue')
plt.plot(t_05, y_05, 'o-', label="Euler's Method, h=0.5", color='red')
plt.plot(t_025, y_025, 's-', label="Euler's Method, h=0.25", color='green')
plt.plot(t_midpoint, y_midpoint, '^-', label="Midpoint Method, h=0.5", color='purple')
plt.title('Solution of the Differential Equation using Various Methods')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.grid(True)
plt.show()
```

### RK4 Method with h = 0.5

```python
# Define the fourth-order Runge-Kutta (RK4) method function
def rk4_method(h, t_final):
    t_values = np.arange(0, t_final + h, h) # t values from 0 to t_final with step h
    y_values = np.zeros(len(t_values)) # Initialize y values array
    y_values[0] = 1 # Initial condition
    
    for i in range(1, len(t_values)):
        k1 = f(t_values[i-1], y_values[i-1])
        k2 = f(t_values[i-1] + h/2, y_values[i-1] + k1*h/2)
        k3 = f(t_values[i-1] + h/2, y_values[i-1] + k2*h/2)
        k4 = f(t_values[i-1] + h, y_values[i-1] + k3*h)
        
        y_values[i] = y_values[i-1] + (h/6) * (k1 + 2*k2 + 2*k3 + k4)
    
    return t_values, y_values

# RK4 method with h = 0.5
t_rk4, y_rk4 = rk4_method(0.5, 2)

# Plotting
plt.figure(figsize=(8, 5))
plt.plot(t, y_analytical, label='Analytical Solution', color='blue')
plt.plot(t_05, y_05, 'o-', label="Euler's Method, h=0.5", color='red')
plt.plot(t_025, y_025, 's-', label="Euler's Method, h=0.25", color='green')
plt.plot(t_midpoint, y_midpoint, '^-', label="Midpoint Method, h=0.5", color='purple')
plt.plot(t_rk4, y_rk4, 'x-', label="RK4 Method, h=0.5", color='orange')
plt.title('Solution of the Differential Equation using Various Methods')
plt.xlabel('t')
plt.ylabel('y(t)')
plt.legend()
plt.grid(True)
plt.show()
```
